```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame()
```

```python
values7 = []
with open("n7.txt", "r") as file:
    for line in file.readlines():
        line = line.replace('\n', '')
        if line.replace('.', '').isnumeric():
            values7.append(float(line))
```

```python
values8 = []
with open("n8.txt", "r") as file:
    for line in file.readlines():
        line = line.replace('\n', '')
        if line.replace('.', '').isnumeric():
            values8.append(float(line))
```

```python
values9 = []
with open("n9.txt", "r") as file:
    for line in file.readlines():
        line = line.replace('\n', '')
        if line.replace('.', '').isnumeric():
            values9.append(float(line))
```

```python
means7 = []
for j in range(9):
    total = 0
    for index in range(10):
        total += values7[index + j*10]
    means7.append(total / 10)
```

```python
means8 = []
for j in range(9):
    total = 0
    for index in range(10):
        total += values8[index + j*10]
    means8.append(total / 10)
```

```python
means9 = []
for j in range(9):
    total = 0
    for index in range(10):
        total += values9[index + j*10]
    means9.append(total / 10)
```

```python
df["K"] = [1, 2, 4, 8, 16, 32, 64, 128, 256]
```

```python
df["10^7"] = means7
df["10^8"] = means8
df["10^9"] = means9
```

```python
plt.figure(figsize=(20, 20))
plt.plot(df["10^7"], df["K"], marker="o", linestyle=":", color="magenta", label="N=10^7")
plt.plot(df["10^8"], df["K"], marker="o", linestyle=":", color="royalblue", label="N=10^8")
plt.plot(df["10^9"], df["K"], marker="o", linestyle=":", color="darkviolet", label="N=10^9")
plt.yticks([1, 2, 4, 8, 16, 32, 64, 128, 256])
plt.xlabel("Tempo médio (segundos)", fontsize=18)
plt.ylabel("Threads", fontsize=18)
plt.legend(fontsize="20")
plt.show()
```
