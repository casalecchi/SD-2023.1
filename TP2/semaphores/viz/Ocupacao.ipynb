```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1_1 = pd.read_csv("ocupacao1_1.csv")
plt.scatter(df1_1["Itens_consumidos"], df1_1["Ocupacao_vetor"], s=2, alpha=0.7, color='green')
plt.xlabel("Itens consumidos")
plt.ylabel("Ocupação do vetor")
plt.title("Vetor com 100 posições - 1 Produtor e 1 Consumidor")
plt.show()
```
